##  Bransley's fern generation using python


![alt text](misc_files/fern.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function_one(x , y):
    '''
    stem generation with a probability of 0.01
    '''
    a = [[0.00 ,0.00 ],[0.00 , 0.16]]
    b = [x , y]
    result = np.dot(a , b)
    return result

def function_two(x , y):
    '''successive smaller leaflets with the probability of 0.85 '''
    a = [[0.85 , 0.04],[-0.04 , 0.85]]
    b = [x , y]
    c = [0.00 , 1.60]
    product = np.dot(a , b)
    result = product + c
    return result

def function_three(x , y):
    '''larger left-hand leaflet with probability of 0.07'''
    a = [[0.20 , -0.26] , [0.23 , 0.22]]
    b = [x , y]
    c = [0.00 , 1.60]
    product = np.dot(a , b)
    result = product + c
    return result

def function_four(x , y):
    '''larger right-hand leaflet with probability of 0.07'''
    a = [[-0.15 , 0.28],[0.26 , 0.24]]
    b = [x,y]
    c = [0.00 , 0.44]
    product = np.dot(a,b)
    result = product + c
    return result



functions = [function_one , function_two , function_three , function_four]
probabilities = [0.01 , 0.85 , 0.07 , 0.07]
total_probability = 0.01 + 0.85 + 0.07 + 0.07  #should be equall to one
```

```python
def bransley_fern_generator(x=0, y=0 , s=0 , points = 100000): 
    x_list = [] 
    y_list = []
    for i in range(points):
        function = np.random.choice(functions , p = probabilities)
        x,y = function(x , y)
        x_list.append(x)
        y_list.append(y)

    plt.scatter(x_list , y_list ,s=s, color = "green")
    return plt.draw()
    
bransley_fern_generator(x,y,0.1,100000)   
```
